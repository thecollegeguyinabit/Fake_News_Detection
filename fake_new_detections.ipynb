```python
# Function to process the texts
def word_cleaner(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)  # Remove text in square brackets
    text = re.sub("\\W", " ", text)  # Remove special characters
    text = re.sub('https?://\S+|www\.\S+', '', text)  # Remove URLs
    text = re.sub('<.*?>+', '', text)  # Remove HTML tags
    text = re.sub('\n', '', text)  # Remove new lines
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)  # Remove punctuation)
    text = re.sub('\w*\d\w*', '', text)  # Remove words containing numbers
    return text
```

```python
df["text"] = df["text"].apply(word_cleaner)
```

```python
x = df["text"]
y = df["label"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
# converting text to vector
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
# logistic regression model

logistic_model = LogisticRegression()
logistic_model.fit(xv_train, y_train)

```

```python
y_pred_logistic = logistic_model.predict(xv_test)
logistic_model.score(xv_test, y_test)
```

```python
print(classification_report(y_test, y_pred_logistic))
```

```python
# Gradient Boosting Classifier
gb_model = GradientBoostingClassifier()
gb_model.fit(xv_train, y_train)
```

```python
pred_gb = gb_model.predict(xv_test)
gb_model.score(xv_test, y_test)
```

```python
# Random Forest Classifier
rf_model = RandomForestClassifier()
rf_model.fit(xv_train, y_train) 
```

```python
pred_rf = rf_model.predict(xv_test)
print(classification_report(y_test, pred_rf))
```

```python

# Collect metrics for each model
metrics = {
    'Accuracy': [round(accuracy_score(y_test, y_pred_logistic),2), round(accuracy_score(y_test, pred_gb),2), round(accuracy_score(y_test, pred_rf),2)],
    'Precision': [round(precision_score(y_test, y_pred_logistic),2), round(precision_score(y_test, pred_gb),2), round(precision_score(y_test, pred_rf),2)],
    'Recall': [round(recall_score(y_test, y_pred_logistic),2), round(recall_score(y_test, pred_gb),2), round(recall_score(y_test, pred_rf),2)],
    'F1 Score': [round(f1_score(y_test, y_pred_logistic),2), round(f1_score(y_test, pred_gb),2), round(f1_score(y_test, pred_rf),2)]
}
model = ['Logistic Regression', 'Gradient Boosting', 'Random Forest']
metrics_df = pd.DataFrame(metrics)
metrics_index = np.arange(len(model))
width = 0.20
multiplier = 0
# Plotting

fig, ax = plt.subplots(layout='constrained')
for attribute, measurement in metrics.items():
    offset = width * multiplier
    rects = ax.bar(metrics_index + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_ylabel('Performance Metrics')
ax.set_title('Model Performance Comparison')
ax.set_xticks(metrics_index + width, model)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 1.5)

plt.show()
```
